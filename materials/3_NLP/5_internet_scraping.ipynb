```python
from twython import Twython
CONSUMER_KEY = "<consumer key>"
CONSUMER_SECRET = "<consumer secret>"
OAUTH_TOKEN = "<application key>"
OAUTH_TOKEN_SECRET = "<application secret>"
twitter = Twython(
    CONSUMER_KEY, CONSUMER_SECRET,
    OAUTH_TOKEN, OAUTH_TOKEN_SECRET)

tweet = twitter.show_status(id=sample_urls['Tweet ID'])
```

```python
def scrape_tweet(url):
    response = requests.get('https://twitter.com/twitter/status/'+url)
    html_string = response.text
    return html_string
```

```python
sample_urls['text'] = sample_urls['Tweet ID'].apply(scrape_tweet)
```
